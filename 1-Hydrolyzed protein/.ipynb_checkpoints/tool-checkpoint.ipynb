### The tool to get the frequency of potential AOP found in multiple proteins

```python
import pandas as pd
file = ".\\ph_A.csv" # stored AOP and frequence in single protein
```

```python
seqdata = list(pd.read_csv(file)["peptide"])  #get the peptide fragment data
freqdata = list(pd.read_csv(file)["frequency"]) #get the frequency of peptide in single protein
count = len(seqdata)
tempseq = sorted(list(set(seqdata))) #de-weight 
```

```python
new_cont = ""
for temp in tempseq:
    tempfreq = 0
    for seq in seqdata:
        if seq == temp:
            currentfreq = int(freqdata[seqdata.index(seq)])
            tempfreq+=currentfreq
    new_cont += temp+","+str(tempfreq)+"\n"
            
```

```python
output="ph_A_stat.csv"
with open(output,"w") as f: #write a file in current path
    f.write("peptide,frequency\n")
    f.write(new_cont)
    f.close()

print("the frequency of AOP existed in all proteins were stored in "+output)
```
