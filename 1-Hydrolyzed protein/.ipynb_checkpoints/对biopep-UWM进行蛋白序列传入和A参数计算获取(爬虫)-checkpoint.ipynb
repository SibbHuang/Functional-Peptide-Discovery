<p>导入相关库：


```python
from selenium import webdriver
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.common.keys import Keys
import time
#为了获取xpath：
from selenium.webdriver.common.by import By
#为了操作下拉列表：
from selenium.webdriver.support.ui import Select

#方便数据处理：
import pandas as pd

#导入四舍五入的库：
import decimal


import smtplib
from email.mime.text import MIMEText
from email.header import Header
import traceback
```

<p>读取csv数据：

```python
protein_csv_location=r".\\PorphyraProtein\\Porphyra.csv"

#将csv的内容读取为一个dataframe，默认第一行为列名：
protein_data=pd.read_csv(protein_csv_location)

#将数据转为列表：
data1=list(protein_data["Sequence"])
data2=list(protein_data["Length"])
```

```python

# executable_path根据情况可写 可不写
driver=webdriver.Firefox()

#使用隐式等待，避免出现定位不到元素：
driver.implicitly_wait(10)

#driver = webdriver.Chrome(executable_path=r"‪D:\anaconda3\envs\python3\geckodriver.exe")

#调整窗口大小
driver.set_window_size(400,640)
driver.get("https://biochemia.uwm.edu.pl/biopep/proteins_data_page1.php")
```

我构建了一个操作的函数：

```python
def into_main_window():
    #点击analysis按钮：
    analysis_button_xpath=r"/html/body/form[1]/input"
    analysis_check=driver.find_element(By.XPATH,analysis_button_xpath).click()
    time.sleep(2)


    #点击calculate按钮：
    calculate_button_xpath=r"/html/body/form/table/tbody/tr[1]/td/input[2]"
    calculate_click=driver.find_element(By.XPATH,calculate_button_xpath).click()
    time.sleep(2)


    #定位下拉列表：
    list_xpath=r"/html/body/form/table/tbody/tr[1]/td[2]/select"
    list_position=driver.find_element(By.XPATH,list_xpath)

    #使用select中的方法定位value的值
    Select(list_position).select_by_value('antioxidative')
    #等待两秒
    time.sleep(2)


    #点击For your Sequence按钮：
    ForYourSequence_button_xpath=r"/html/body/form/table/tbody/tr[2]/td/table/tbody/tr[2]/td/div/input"

    ForYourSequence_click=driver.find_element(By.XPATH,ForYourSequence_button_xpath).click()

    time.sleep(2)

#给表格传入蛋白序列：
def get_result(data):
    # python传入的值
    test_txt = data
    text_area_xpath=r"/html/body/form/div/table/tbody/tr/td[2]/div/textarea"
    driver.find_element(By.XPATH,text_area_xpath).send_keys(test_txt)

    # 拼接js
    #js = 'var txt = document.getElementById("textarea"); txt.value="'+test_txt+'";'  
    # 执行js代码
    #driver.execute_script(js)

    time.sleep(2)

    #点击report：
    report_button_xpath=r"/html/body/form/div/table/tbody/tr/td[3]/div/input"
    report_click=driver.find_element(By.XPATH,report_button_xpath).click()
    time.sleep(3)


    #获取A参数：
    Aparameters_xpath=r"/html/body/table/tbody/tr[2]/td[4]"
    Aparameters_value=driver.find_element(By.XPATH,Aparameters_xpath).text

    #返回上一个界面：
    driver.back()
    time.sleep(2)
    driver.refresh()
    time.sleep(1)
    return Aparameters_value
```

```python
#写入文件的位置：
file_location=r".\\PorphyraProtein\\A_result.csv"

# executable_path根据情况可写 可不写
driver=webdriver.Firefox()
#使用隐式等待，避免出现定位不到元素：
driver.implicitly_wait(10)

#driver = webdriver.Chrome(executable_path=r"‪D:\anaconda3\envs\python3\geckodriver.exe")
#调整窗口大小
driver.set_window_size(400,640)
driver.get("https://biochemia.uwm.edu.pl/biopep/proteins_data_page1.php")

#先进入主要的页面：
into_main_window()

count=len(data1)

with open(file_location,"a+") as f:

    f.write("Sequence,Length,A,biopeptide_num\r")
    for i in range(count):
        try:
            A_param=get_result(data1[i])
            num=i+1
            #计算活性肽的数目：
            biopeptide_num=float(A_param)*float(data2[i])
            biopeptide_num=str(decimal.Decimal(biopeptide_num).quantize(decimal.Decimal("0")))
            f.write(data1[i]+","+str(data2[i])+","+A_param+","+biopeptide_num+"\r")
            print("finished\t"+str(round(num/count*100,3))+"%")


        except Exception as e:
#             error_message = str(e)+"\r代码又出错了，不过还是能运行"
#             send_email('1821367541@qq.com', error_message)
            print("Sequence: "+data1[i] +"can't calculate successfully.")
            driver.back()
        continue
    f.close()
#     send_email('1821367541@qq.com', "好像顺利完成了")
       
        
    
```
