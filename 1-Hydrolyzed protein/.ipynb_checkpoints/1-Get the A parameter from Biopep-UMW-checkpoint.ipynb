### Function to get A parameter  automatically:

```python
def into_main_window(driver):
    """
    a sample function to enter the submit page
    return None
    """
    #going to the BIOPEP-UMW
    driver.get(url = "https://biochemia.uwm.edu.pl/biopep/start_biopep.php")
    time.sleep(1)
    #click the analysis button:
    analysis_button_xpath=r"/html/body/form[1]/input"
    analysis_check=driver.find_element(By.XPATH,analysis_button_xpath).click()
    time.sleep(2)

    #click the calculate button：
    calculate_button_xpath=r"/html/body/form/table/tbody/tr[1]/td/input[2]"
    calculate_click=driver.find_element(By.XPATH,calculate_button_xpath).click()
    time.sleep(2)

    #locate the drop-down list:
    list_xpath=r"/html/body/form/table/tbody/tr[1]/td[2]/select"
    list_position=driver.find_element(By.XPATH,list_xpath)

    #locate the element in list named antioxidative:
    Select(list_position).select_by_value('antioxidative')
    time.sleep(2)

    #locate the element called For your sequence:
    ForYourSequence_button_xpath=r"/html/body/form/table/tbody/tr[2]/td/table/tbody/tr[2]/td/div/input"
    ForYourSequence_click=driver.find_element(By.XPATH,ForYourSequence_button_xpath).click()
    time.sleep(2)
    
    return None


def get_result(data):
    """
    A function to send your sequence into box
    data: the sequence was submitted
    return: A parameter for this sequence
    """
    test_txt = data
    text_area_xpath=r"/html/body/form/div/table/tbody/tr/td[2]/div/textarea"
    driver.find_element(By.XPATH,text_area_xpath).send_keys(test_txt)
    time.sleep(2)

    #click report button:
    report_button_xpath=r"/html/body/form/div/table/tbody/tr/td[3]/div/input"
    report_click=driver.find_element(By.XPATH,report_button_xpath).click()
    time.sleep(3)
    
    #get the A parameter:
    Aparameters_xpath=r"/html/body/table/tbody/tr[2]/td[4]"
    Aparameters_value=driver.find_element(By.XPATH,Aparameters_xpath).text
    
    #back to last page:
    driver.back()
    time.sleep(2)
    driver.refresh()
    time.sleep(1)
    return Aparameters_value

def start(): 
    """
    Launch your firefox browser
    return driver
    """
    driver=webdriver.Firefox()
    driver.implicitly_wait(10)
    #set the window size:
    driver.set_window_size(300,600)
    return driver

def restart(driver): 
    """
    treat with error situation
    return None
    """
    into_main_window(driver)
    driver.refresh
    time.sleep(1)
    return None
```

### Main section:

```python
error_seq = ""
#it can record the sequence which happened fatal error ant can't get A parameter
count=len(data1)
# a lenght of your sequence's number

if __name__ == "__main__":
    #enter the main working page:
    driver = start()
    into_main_window(driver)
    #got and stored the A parameter:
    with open(file_location,"a+") as f:
        f.write("Sequence,Length,A,biopeptide_num\r")
        for i in range(count):
            try:
                A_param=get_result(data1[i])
                num=i+1
                biopeptide_num=float(A_param)*float(data2[i])
                biopeptide_num=str(decimal.Decimal(biopeptide_num).quantize(decimal.Decimal("0")))
                f.write(data1[i]+","+str(data2[i])+","+A_param+","+biopeptide_num+"\r")
                print("finished\t"+str(round(num/count*100,3))+"%")
            except Exception as e:
                error_seq += str(i+1) + "," + data1[i] + "\n"
                restart(driver)
                time.sleep(1)
            continue
        f.close()

    #to record the possible sequence which not get the A parameter:
    with open(error_seq_location,"w+") as f:
        f.write("No,Sequence\n")
        f.write(error_seq)
        f.close()
```
