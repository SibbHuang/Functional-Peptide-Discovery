 ## Detect the Special Amino Acid in peptide set：


```python
import pandas as pd

def collect_aminoacid_from_peptide(peptide):
    """
    get the amino acid list
    return list
    """
    aalist=[aminoacid for aminoacid in peptide]
    del peptide
    return aalist

def search_hydrophobic_amino_acid(aalist):
    """
    find the hydrophobic amino acid existed in peptide such as Leu, Val, Ala, Pro and Phe
    return count, int object represent the number of hydrophobic amino acid
    """
    count=0
    for aminoacid in aalist:
        if aminoacid == "L":
            count+=1
        elif aminoacid == "V":
            count+=1
        elif aminoacid == "A":
            count+=1
        elif aminoacid == "P":
            count+=1
        elif aminoacid == "F":
            count+=1
        else:
            pass
    return count

def search_aromatic_amion_acid(aalist):
    """
    find the count of aromatic amino acid in peptide such as Trp, Tyr and Phe
    return count, int object
    """
    count=0
    for aminoacid in aalist:
        if aminoacid == "W":
            count+=1
        elif aminoacid == "Y":
            count+=1
        elif aminoacid == "F":
            count+=1
        else:
            pass
    return count



def search_sulfur_containing_amino_acid(aalist):
    """
    find the count of amino acid with S element
    return count
    """
    count=0
    for aminoacid in aalist:
        if aminoacid == "C":
            count+=1
        elif aminoacid == "M":
            count+=1
        else:
            pass
    return count

def search_basic_amino_acid(aalist):
    """
    find the count of base amino acid in peptide,such as His, Lys and Arg
    return count
    """
    count=0
    for aminoacid in aalist:
        if aminoacid == "H":
            count+=1
        elif aminoacid == "K":
            count+=1
        elif aminoacid == "R":
            count+=1
        else:
            pass
    return count

def collectpeptidescore(score,peptide,smile):
    """
    evalute the special AA 
    """
    if score == 0:
        judgement = "-"
    if score != 0:
        judgement = ""
        for i in range(score):
            judgement += "+"
    peptideresult = peptide+","+judgement+","+smile
    return peptideresult

def fromtxtreadpeptide(filename):
    with open(filename,"r+",encoding="utf8") as f:
        data=f.readlines()  #读取整个文件
        data=[string.replace("\n","") for string in data]
        f.close()
    return data

def loopcheck(peptideset,smileset,output="saaResult.csv"):
    with open(output,"w") as f:
        f.write("peptide,score,smiles\n")
        for index in range(len(peptideset)):
            peptidescore = 0
            aalist = collect_aminoacid_from_peptide(peptideset[index])
            count = search_hydrophobic_amino_acid(aalist)
            if count != 0:
                peptidescore+=1
            
            count = search_basic_amino_acid(aalist)
            if count != 0:
                peptidescore+=1
            
            count = search_sulfur_containing_amino_acid(aalist)
            if count != 0:
                peptidescore+=1
            
            count = search_aromatic_amion_acid(aalist)
            if count != 0:
                peptidescore+=1
            
            peptideresult = collectpeptidescore(peptidescore,peptideset[index],smileset[index])    
            f.write(peptideresult+"\n")
        f.close()
        return None
```

```python
file = "example.csv" #input file
df = pd.read_csv(file)
```

```python
peptideset = df["peptide"]
smileset = df["smile"]
output="ExampleSaaResult.csv"
loopcheck(peptideset,smileset,output="ExampleSaaResult.csv")
print("Luckly! The result were stored in "+output)

#check result:
dfresult = pd.read_csv(output)
dfresult.head()
```

### Find the amino acid with Aromatic ring:

```python
import pandas as pd
def findAAA(seq):
    """
    find the count of aromatic amino acid in peptide
    return text, string object containing the peptide sequence and frequency of aromatic amino acid
    """
    Pro_count = 0
    for i in seq:
        if i == "P":
            Pro_count += 1
            break
        else:
            pass
    Tyr_count = 0 
    for i in seq:
        if i == "Y":
            Tyr_count += 1
            break
        else:
            pass
    Phe_count = 0
    for i in seq:
        if i == "F":
            Phe_count += 1
            break
        else:
            pass
    text = seq+","+str(Pro_count)+","+str(Tyr_count)+","+str(Phe_count)+"\n"
    return text

def loop(data,fn="specialAA.csv"):
    """
    treat with multiple peptides
    return none
    """
    with open (fn,"w") as f:
        f.write("pepname,Pro,Tyr,Phe\n")
        for seq in data:
            text=findAAA(seq)
            f.write(text)
        f.close()
    return None
```

```python
p = "example.csv"
data = pd.read_csv(p)["peptide"]
output="specialAAA.csv"
loop(data,fn=output)
print("Luckly! Data were stored in "+output)
```
